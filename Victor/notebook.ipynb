Problématique du projet : Quels sont les facteurs qui influencent les ventes et comment prédire leur impact ?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
pd.set_option("display.max_rows", None)
```

```python
train= pd.read_csv('train.csv')
test= pd.read_csv('test.csv')
features= pd.read_csv('features.csv')
stores= pd.read_csv('stores.csv')

train['Date']= pd.to_datetime(train['Date'])
test['Date']= pd.to_datetime(test['Date'])
features['Date']= pd.to_datetime(features['Date'])
```

```python
features.head()
```

```python
weekly_sales_by_date = train.groupby('Date')['Weekly_Sales'].sum()

plt.figure(figsize=(12, 6))  # Définissez la taille du graphique
plt.plot(weekly_sales_by_date, marker='o', linestyle='-')
plt.title('Ventes Hebdomadaires en Fonction de la Date')
plt.xlabel('Date')
plt.ylabel('Ventes Hebdomadaires')
plt.grid(True)

# Affichez le graphique
plt.show()
```
